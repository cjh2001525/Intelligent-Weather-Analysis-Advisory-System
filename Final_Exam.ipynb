````python
"""
Corrected task manager (beginner level).
Handwritten final version with tiny doctests for helpers.
Run tests: python -m doctest -v this_file.py
"""

def format_tasks(tasks):
    """
    Return a numbered list as a string, or '(no tasks)' if empty.

    >>> format_tasks([])
    '(no tasks)'
    >>> format_tasks(['a','b']).splitlines()[0]
    '1. a'
    """
    if not tasks:
        return "(no tasks)"
    lines = []
    for i, t in enumerate(tasks, start=1):
        lines.append(f"{i}. {t}")
    return "\n".join(lines)


def add_task(tasks, text):
    """
    Add a non-empty trimmed task; return True if added, False otherwise.

    >>> items = []
    >>> add_task(items, "  read  ")
    True
    >>> items
    ['read']
    >>> add_task(items, "   ")
    False
    """
    t = (text or "").strip()
    if not t:
        return False
    tasks.append(t)
    return True


def remove_task(tasks, idx):
    """
    Remove task by 1-based index; return removed text or None when invalid.

    >>> items = ['a','b','c']
    >>> remove_task(items, 2)
    'b'
    >>> remove_task(items, 5) is None
    True
    """
    if not isinstance(idx, int):
        return None
    if 1 <= idx <= len(tasks):
        return tasks.pop(idx - 1)
    return None


def count_tasks(tasks):
    """
    Return the number of tasks.

    >>> count_tasks(['x','y'])
    2
    >>> count_tasks([])
    0
    """
    return len(tasks)


def display_menu():
    """Print menu options (interactive; not unit-tested)."""
    print("1. Add task")
    print("2. List tasks")
    print("3. Remove task")
    print("4. Count tasks")
    print("5. Quit")


def get_choice():
    """Read and normalise a menu choice string."""
    return input("Choose (1-5): ").strip()


def main():
    """
    Main loop (beginner-level only: lists, loops, minimal try/except).
    """
    tasks = []
    while True:
        display_menu()
        choice = get_choice()

        if choice == "1":
            text = input("Task: ")
            if add_task(tasks, text):
                print("Added.")
            else:
                print("Empty task ignored.")
            print(f"Total tasks: {count_tasks(tasks)}")

        elif choice == "2":
            print(format_tasks(tasks))
            print(f"Total tasks: {count_tasks(tasks)}")

        elif choice == "3":
            if not tasks:
                print("Nothing to remove.")
                continue
            print(format_tasks(tasks))
            raw = input("Index to remove: ").strip()
            try:
                idx = int(raw)
            except Exception:
                print("Please enter a number.")
                continue
            removed = remove_task(tasks, idx)
            if removed is None:
                print("Index out of range.")
            else:
                # Optional confirmation pattern (kept minimal and inline)
                print(f"Removed: {removed}")
            print(f"Total tasks: {count_tasks(tasks)}")

        elif choice == "4":
            print(f"Total tasks: {count_tasks(tasks)}")

        elif choice == "5":
            print("Bye.")
            break

        else:
            print("Invalid choice. Enter 1-5.")


if __name__ == "__main__":
    import doctest
    doctest.testmod()
    # Uncomment to run interactively:
    # main()
````

## 4.1 Manual Implementation (15 marks)

Implement three core functions at an introductory level only. Use `print()`/`input()`, `while` loops, simple conditionals, and lists. Keep helpers tiny and add **doctests** where possible. Interactive functions can include `# doctest: +SKIP`.

Functions to provide:
- `display_menu()` — print the menu (interactive; not unit-tested)
- `get_user_choice()` — read and basic-validate the choice (returns a string '1'..'5')
- `main()` — programme loop that uses a list in memory and the two helpers above

> Note: Do not add advanced libraries or file persistence here. Keep everything within Weeks 2–3 & 6 (basic I/O and minimal validation).


````python
# S4 — beginner-level manual implementation

def display_menu():
    """Print menu options.

    >>> display_menu()  # doctest: +SKIP
    1. Add task
    2. List tasks
    3. Remove task
    4. Count tasks
    5. Quit
    """
    print("1. Add task")
    print("2. List tasks")
    print("3. Remove task")
    print("4. Count tasks")
    print("5. Quit")


def get_user_choice():
    """Get and normalise a user choice string '1'..'5' (basic validation).

    >>> # Example (interactive): choice = get_user_choice()  # doctest: +SKIP
    """
    choice = input("Choose (1-5): ").strip()
    # keep validation minimal and friendly
    if choice not in {"1", "2", "3", "4", "5"}:
        print("Invalid choice. Enter 1-5.")
        return ""  # caller may loop again
    return choice


def _format_tasks(tasks):
    """Return a printable numbered view for a task list.

    >>> _format_tasks([])
    '(no tasks)'
    >>> _format_tasks(['alpha','beta']).splitlines()[1]
    '2. beta'
    """
    if not tasks:
        return "(no tasks)"
    return "\n".join(f"{i}. {t}" for i, t in enumerate(tasks, start=1))


def main():
    """Main loop using only beginner patterns (lists, while, input/print).

    >>> main()  # doctest: +SKIP
    """
    tasks = []
    while True:
        display_menu()
        choice = get_user_choice()
        if choice == "":
            continue  # invalid input; re-display menu

        if choice == "1":  # Add
            text = input("Task: ").strip()
            if text:
                tasks.append(text)
                print("Added.")
            else:
                print("Empty task ignored.")
            print(f"Total tasks: {len(tasks)}")

        elif choice == "2":  # List
            print(_format_tasks(tasks))
            print(f"Total tasks: {len(tasks)}")

        elif choice == "3":  # Remove
            if not tasks:
                print("Nothing to remove.")
                continue
            print(_format_tasks(tasks))
            raw = input("Index to remove: ").strip()
            # minimal safe conversion
            if not raw.isdigit():
                print("Please enter a number.")
                continue
            idx = int(raw)
            if 1 <= idx <= len(tasks):
                removed = tasks.pop(idx - 1)
                print(f"Removed: {removed}")
            else:
                print("Index out of range.")
            print(f"Total tasks: {len(tasks)}")

        elif choice == "4":  # Count
            print(f"Total tasks: {len(tasks)}")

        elif choice == "5":  # Quit
            print("Bye.")
            break
````
